```python
personas = []
continuar = True

print("=== INICIANDO SISTEMA ===")

# Entrada y procesamiento de datos
while continuar:

    nombre = input("Escriba su nombre si desea continuar. Caso contrario escriba 'cancelar': ")

    if nombre in ["cancelar", "Cancelar", "CANCELAR"]:
        continuar = False # el programa se detendrá
        break
    else:
        edad= int(input(f"¿Cuál es tu edad {nombre}? "))
        nota= int(input(f"¿Cuál es tu nota {nombre}? "))

        # append es el equivalente a push en Python
        personas.append([nombre, edad, nota])

# Salida de datos
print("\n--- MATRIZ GENERAL ---")
for persona in personas:
    nombre = persona[0]
    edad = persona[1]
    nota = persona[2]

    print(f"Nombre: {nombre}, Edad: {edad}, Nota: {nota}")

print("\n--- PERSONAS ORDENADAS POR NOTA DE MAYOR A MENOR ---")
# list() hace exactamente lo mismo que slice() en JS: crea una copia independiente
personas_ordenadas = list(personas)

# Creamos la función que le dice a Python qué índice mirar para ordenar
def comparar_notas(elemento):
    return elemento[2]

# Aplicamos el ordenamiento de mayor a menor (reverse=True)
personas_ordenadas.sort(key=comparar_notas, reverse=True)

for persona in personas_ordenadas:
    print(f"Nombre: {persona[0]} | Edad: {persona[1]} | Nota: {persona[2]}")

print("\n--- PROMEDIO GENERAL ---")
suma = 0
for persona in personas:
    suma = suma + persona[2] # Vamos acumulando las notas

# len() reemplaza a .length de JavaScript
if len(personas) > 0:
    promedio = suma / len(personas)
    print(f"Promedio del grupo: {promedio}")
else:
    print("No se ingresaron datos para promediar.")
```
